```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# total_bill의 중간값 이하인 것만 조회
tips.loc[tips["total_bill"] <= tips["total_bill"].median(), :]
```

```python
# day가 Sun인 데이터만 조회
tips.loc[tips['day'] == 'Sun', :]

# tip의 평균보다 크면서, time이 Dinner인 데이터만 조회
# AND 조건, A이고 B이다. A이면서 B이다
# OR 조건, A 또는 B이다

#pandas의 다중조건 행 추출
# df.loc[(조건1) & (조건2), :]
# &, AND 조건 / |, OR 조건
tips.loc[(tips['tip'] > tips['tip'].mean()) & (tips['time'] == 'Dinner'), :]
```

```python
tips.loc[(tips['tip'] > tips['tip'].mean()) &
    (tips['time'] == 'Dinner') &
    (tips['size'] == 3) &
    (tips['sex'] == 'Male') &
    (tips['day'] == 'Sat'),
:]
```
